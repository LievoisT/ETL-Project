```python
import pandas as pd
import os
```

```python
file = os.path.join("vgsales.csv")
df = pd.read_csv(file)
```

```python
to_drop = ['NES', 'SNES', 'SAT', '2600', 'GB', 'GEN', 'NG', 'SCD', 'WS', '3DO', 'TG16', 'GG', 'PCFX']
df = df[~df['Platform'].isin(to_drop)]
df = df.dropna()
df
```

```python
df = df.drop_duplicates(subset=['Name', 'Platform'], keep='first')
```

```python
title_df = df[["Name", "Year", "Publisher", 'Platform', 'Genre']]
title_df = title_df.rename(columns={"Name": "game_title", "Publisher": "publisher", "Year":"year_released",\
                                    'Genre':'genre'})
title_df = title_df.dropna(subset=['game_title'])
title_df
```

```python
title_df = title_df.drop_duplicates(subset=['game_title'])
title_df
```

```python
title_df = title_df.drop(columns=['Platform'])
title_df = title_df.reset_index(drop=True)
title_df.index.name = "id"
title_df
```

```python
genre_df = title_df[['genre']]
genre_df.index.name = "title_id"
genre_df.to_csv('genre.csv')
genre_df
```

```python
title_df.to_csv('title.csv')
```

CREATE SALES DATAFRAME

```python
sales_df = df
```

```python
sales_df = sales_df[['Name', 'Platform','NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_df
```

```python
title_id_dict = dict(zip(title_df.game_title, title_df.index))
sales_df['Name'] = sales_df['Name'].map(title_id_dict)
sales_df
```

```python
sales_df = sales_df.rename(columns={"Name": "title_id", "NA_Sales": "na_sales", "EU_Sales":"eu_sales",\
                                    'JP_Sales':'jp_sales', 'Other_Sales':'other_sales', 'Platform':"platform_id"})
sales_df = sales_df.set_index('title_id')
```

```python
sales_df.to_csv("sales.csv")
```
